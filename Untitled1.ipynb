```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

MY data
**Here i will upload my data **

```python
mc=pd.read_csv('/content/Mall_Customers2.csv')
```

```python
mc.head()
```

```python
sns.heatmap(mc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
mc.dropna(inplace=True)
sns.heatmap(mc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

#encoding

```python
gender=pd.get_dummies(mc["Gender"],drop_first=True)
```

```python
mc.drop(['Gender','CustomerID'],inplace=True,axis=1)
```

```python
mc=pd.concat([mc , gender],axis=1)
```

```python
sns.heatmap(mc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
mc.head()
```

```python
# Create income bins
income_bins = [0, 30, 60, 90, 120, 150]
income_labels = ['0-30', '31-60', '61-90', '91-120', '121-150']
mc['Income_Group'] = pd.cut(mc['Annual Income (k$)'], bins=income_bins, labels=income_labels, right=False)

# Create age bins
age_bins = [0, 20, 30, 40, 50, 60, 70]
age_labels = ['0-19', '20-29', '30-39', '40-49', '50-59', '60-69']
mc['Age_Group'] = pd.cut(mc['Age'], bins=age_bins, labels=age_labels, right=False)

# Pie chart for Male distribution within Income Groups
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
male_income_distribution = mc.groupby('Income_Group')['Male'].value_counts(normalize=False).unstack().fillna(0)
male_income_distribution.plot(kind='pie', y=True, autopct='%1.1f%%', legend=False, ax=plt.gca())
plt.title('Male Distribution within Income Groups')
plt.ylabel('')

# Pie chart for Male distribution within Age Groups
plt.subplot(1, 2, 2)
male_age_distribution = mc.groupby('Age_Group')['Male'].value_counts(normalize=True).unstack().fillna(0)
male_age_distribution.plot(kind='pie', y=True, autopct='%1.1f%%', legend=False, ax=plt.gca())
plt.title('Male Distribution within Age Groups')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
